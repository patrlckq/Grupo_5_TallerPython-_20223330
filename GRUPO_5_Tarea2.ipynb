```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Ejercicio 1
data_ciudades = { 'Ciudad': ['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Chiclayo', 'Piura', 'Iquitos', 'Huancayo', 'Tacna', 'Pucallpa'],
                  'Habitantes': [1047796, 100169, 92331, 428450, 305717, 484475, 441649, 385098, 294395, 289734],
                  'Area_Km2': [2672.28, 1545.77, 1487.7, 116.5, 279.89, 6217.26, 368.9, 109.19, 59.4, 483.44],
                  'Altitud_m': [154, 2325, 34, 3399, 29, 29, 106, 3271, 562, 156],
                  'Densidad_poblacion_por_km2': [3924, 64.8, 62.1, 3673, 1091.6, 77.9, 1196, 3528.6, 4966, 587.2] }
df_nf = pd.DataFrame(data_ciudades)
```

```python
#Ejercicio 2
data_ciudades_2 = { 'Ciudad': ['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Chiclayo', 'Piura', 'Iquitos', 'Huancayo', 'Tacna', 'Pucallpa'],
                    'Gentilicio': ['Limense', 'Arequipeño', 'Trujillano', 'Cusqueño', 'Chiclayano', 'Piurano', 'Iquiteño', 'Huancaino', 'Tacneño', 'Pucallpino'],
                    'Provincia': ['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Chiclayo', 'Piura', 'Maynas', 'Huancayo', 'Tacna', 'Coronel Portillo'],
                    'Region': ['Lima', 'Arequipa', 'La Libertad', 'Cusco', 'Lambayeque', 'Piura', 'Loreto', 'Junín', 'Tacna', 'Ucayali']}
df_nombres = pd.DataFrame(data_ciudades_2)
```

```python
#Ejercicio 3
df_cuadro_1 = pd.merge(df_nf, df_nombres, on='Ciudad', how='inner')
df_cuadro_1
```

```python
# Pregunta 4

fig, ax = plt.subplots(figsize = (12,6))
ax.bar( x = df_cuadro_1["Ciudad"], height = df_cuadro_1["Habitantes"], color = "purple")
ax.set_title("Número de habitantes por ciudad en Perú")
ax.set_xlabel("Ciudades")
ax.set_ylabel("Número de habitantes")

plt.show()
```

```python
# Pregunta 5

fig, ax = plt.subplots(figsize = (12,6))
ax.scatter( x = df_cuadro_1["Altitud_m"], y = df_cuadro_1["Habitantes"], color = "orange")
ax.set_title("Relación entre altitud y número de habitantes de las ciudades del Perú")
ax.set_xlabel("Altitud en metros")
ax.set_ylabel("Habitantes")
ax.grid(True, alpha = 0.2)

plt.show()
```
